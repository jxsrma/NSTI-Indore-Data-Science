```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_global_water_consumption.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
totalWC = df[['Country','Total Water Consumption (Billion Cubic Meters)']].groupby('Country').sum()
totalWC
```

```python
sns.barplot(data=totalWC,x = 'Country',y='Total Water Consumption (Billion Cubic Meters)', gap= 0.5)
```
