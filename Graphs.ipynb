```python

# Sample data
hours_studied = [2, 4, 6, 8, 10, 12, 14]
scores = [50, 60, 65, 70, 85, 90, 95]

# Scatter plot
plt.scatter(hours_studied, scores, color='blue', label='Student Data')
plt.title('Scatter Plot: Hours Studied vs Score')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate large dataset with NumPy
np.random.seed(42)  # For reproducibility
hours_studied = np.random.uniform(1, 15, 1000)  # 1000 random values between 1 and 15
scores = 40 + (hours_studied * 3) + np.random.normal(0, 5, 1000)  # Linear relation with some noise

# Scatter plot
plt.scatter(hours_studied, scores, color='blue', alpha=0.5, label='Student Data')
plt.title('Scatter Plot: Hours Studied vs Score (Large Dataset)')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate stock market dataset
np.random.seed(42)  # For reproducibility
days = 100  # Number of days
companies = ['Company A', 'Company B', 'Company C', 'Company D', 'Company E']

# Simulate percentage changes in stock prices for each company
stock_data = {company: np.random.normal(0, 2, days) for company in companies}

# Prepare data for box plot
data = [stock_data[company] for company in companies]

# Box plot
plt.boxplot(data, labels=companies, patch_artist=True)
plt.title('Box Plot: Stock Market Daily Percentage Changes')
plt.ylabel('Daily % Change')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
