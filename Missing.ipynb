```python
import pandas as pd
import numpy as np
```

```python


# Creating a sample DataFrame with missing values
data = {
    'ID': range(1, 21),  # Student IDs from 1 to 20
    'Name': ['John', 'Alice', 'Bob', 'Charlie', np.nan, 'Eve', 'Frank', 'Grace', 'Hannah', 'Ian',
             'Jack', 'Kate', 'Leo', 'Mona', 'Nathan', np.nan, 'Olivia', 'Peter', 'Quinn', 'Rachel'],
    'Age': [20, 21, np.nan, 19, 22, 21, np.nan, 20, 19, 23, 22, 21, np.nan, 20, 19, 22, 23, 21, 20, np.nan],
    'Marks': [85, 78, np.nan, 90, 88, 76, 95, np.nan, 80, 70, 85, np.nan, 89, 92, 77, 88, 90, 85, np.nan, 75],
    'Attendance': [90, 85, 80, 75, np.nan, 88, 92, 85, np.nan, 89, 87, 91, 80, np.nan, 95, 82, 88, np.nan, 85, 90]
}

df = pd.DataFrame(data)

# Display DataFrame
df
```

```python
df['Marks'].mode()
```

```python
# Identify the missing values in the dataset.
df.isna().sum()
```

```python
# Fill the missing Names with "Unknown".
df['Name'].fillna('Unknown')
```

```python
# Fill the missing Age values with the mean age.
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
# Fill the missing Marks values with the median marks.
df['Marks'].median()
df['Marks'] = df['Marks'].fillna(df['Marks'].median())
df
```

```python
# Fill the missing Attendance values using forward fill (ffill).
df['Attendance'] = df['Attendance'].fillna(df['Attendance'].ffill())
df
```

```python
# Fill the missing Attendance values using backward fill (bfill).
df['Attendance'] = df['Attendance'].fillna(df['Attendance'].bfill())
df
```

```python
# Fill the missing Marks with 0.
df['Marks'].fillna(0)
```

```python
# Drop rows with missing values.
df.drop(df.isna())
```

```python
# Drop columns with missing values.
df.drop(df.isna(),axis=1)
```

```python

# Replace missing values with the most frequent value (mode).
df['Marks'].fillna(df['Marks'].mode())
```
